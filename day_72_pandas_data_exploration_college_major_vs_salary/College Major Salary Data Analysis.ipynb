```python
import pandas as pd

pd.options.display.float_format = "{:,.2f}".format
```

```python
df = pd.read_csv("salaries_by_college_major.csv")
df.head()
```

```python
df[df.isnull().any(axis=1)]  # check at least one True (NaN) in row
```

```python
df.dropna(axis=0, inplace=True)
df[df.isnull().any(axis=1)]
```

```python
df.columns
```

```python
df.iloc[df["Mid-Career Median Salary"].idxmax()]
```

```python
df.iloc[df["Mid-Career Median Salary"].idxmin()]
```

```python
spread_col = df["Mid-Career 90th Percentile Salary"].subtract(
    df["Mid-Career 10th Percentile Salary"]
)  # alternatively: df[col_A] - df[col_B]
df.insert(1, "Spread", spread_col)
df.head()
```

```python
# Low Risk
df.sort_values(
    by="Spread", ascending=True
)[["Undergraduate Major", "Spread"]].head()
```

```python
# Highest Potential
mid_90_col = "Mid-Career 90th Percentile Salary"
df.sort_values(by=mid_90_col, ascending=False)\
    [["Undergraduate Major", mid_90_col]].head()
```

```python
# Highest Spread
df.sort_values(
    by="Spread", ascending=False
)[["Undergraduate Major", "Spread"]].head()
```
